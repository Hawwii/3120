```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
data= pd.read_csv('Spotify_Youtube.csv')
data.head()
data.shape
```

```python
artist_streams = data.groupby('Artist')['Stream'].sum()
sorted_artists = artist_streams.sort_values(ascending=False)
top_10_artists = sorted_artists.head(10).index.tolist()
filtered_df = data[data['Artist'].isin(top_10_artists)]
filtered_df.shape
filtered_df

top_artists = artist_streams.sort_values(ascending=False)
```

```python


artist_counts = data.groupby('Artist')['Album_type'].nunique()

artists_with_both = artist_counts[artist_counts == 2].index.tolist()

df_filtered = data[data['Artist'].isin(artists_with_both)]

artist_streams = df_filtered.groupby('Artist')['Stream'].sum()

top_artists = ['Ed Sheeran', 'Dua Lipa', 'XXXTENTACION', 'Justin Bieber', 'Khalid', 'Ariana Grande', 'Billie Eilish', 'The Chainsmokers', 'Shawn Mendes', 'Calvin Harris']

for artist in top_artists:
    artist_df_original = data[data['Artist'] == artist]
    print(artist_df_original.shape)
    

    
    
```

```python
# Group by artist and calculate the number of unique album types for each artist
artist_counts = data.groupby('Artist')['Album_type'].nunique()

# Select artists with both singles and albums
artists_with_both = artist_counts[artist_counts == 2].index.tolist()

# Filter the original dataset to include only the rows corresponding to the selected artists
df_filtered = data[data['Artist'].isin(artists_with_both)]

# Display the resulting DataFrame with all songs by the top 10 artists that have both singles and albums
df_filtered.head(30)

# Define the top 10 artists
top_artists = ['Ed Sheeran', 'Dua Lipa', 'XXXTENTACION', 'Justin Bieber', 'Khalid', 'Ariana Grande', 'Billie Eilish', 'The Chainsmokers', 'Shawn Mendes', 'Calvin Harris']

# Create a dictionary to store DataFrames for each top artist
artist_dfs = {}

# Iterate over the top artists
for artist in top_artists:
    # Filter the dataset for the current artist
    artist_df = df_filtered[df_filtered['Artist'] == artist]
    
    # Store the DataFrame in the dictionary
    artist_dfs[artist] = artist_df

combined_df = pd.concat(artist_dfs.values(), ignore_index=True)
```
